```python
from bs4 import BeautifulSoup
import requests
import csv
import pandas as pd
import ssl
ssl._create_default_https_context = ssl._create_unverified_context
```

```python
url = 'http://tuvalum.com/bicicletas-carretera'
```

```python
response = requests.get(url, 'html.parser')
print(response)
```

```python
soup = BeautifulSoup(response.content)
soup
```

```python
script_tag = soup.find('script', type='application/ld+json')
script_tag
```

```python
import json
```

```python
json_data = json.loads(script_tag.string)
json_data
```

```python
nombres = []
precios = []

for bicicleta in json_data['itemListElement']:
    nombre= bicicleta['name']
    precio = bicicleta['offers']['price']
    nombres.append(nombre)
    precios.append(precio)

print(nombres)  
print('--------')
print(precios)
```
