```python
datos_tour=[]
datos_giro=[]
datos_vuelta=[]

for filaA in filas_grandes:
    if 'Tour' in filaA.text:
        if filaA.find('b') is not None:
            datos_tour.append(filaA.find('b').text.strip())
        else:
            datos_tour.append('Nan')

    elif 'Giro' in filaA.text:
        if filaA.find('b') is not None:
            datos_giro.append(filaA.find('b').text.strip())
        else:
            datos_giro.append('Nan')
    else:
        if filaA.find('b') is not None:
            datos_vuelta.append(filaA.find('b').text.strip())
        else:
            datos_vuelta.append('Nan')

print(datos_tour)
print('--------')
print(datos_giro)
print('--------')
print(datos_vuelta)
```
