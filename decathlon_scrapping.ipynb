```python
from bs4 import BeautifulSoup
import requests
import csv
import pandas as pd
import ssl
ssl._create_default_https_context = ssl._create_unverified_context
```

```python
url = 'https://www.decathlon.es/es/deportes/ciclismo/bicicletas-carretera-de-carbono'
```

```python
response = requests.get(url, 'html.parser')
print(response)
```

```python
soup = BeautifulSoup(response.content)
soup
```

```python
modelos = soup.find_all('a', class_="vtmn-block vtmn-w-full vtmn-no-underline gt-small-desktop:vtmn-leading-normal vtmn-uppercase product-title")
modelos
```

```python
lista_modelos = []

for modelo in modelos:
    nombre = modelo.text.strip().split('\n')[1].replace('Segunda Vida - ','')
    lista_modelos.append(nombre)

lista_modelos
```

```python
precios = soup.find_all('div', class_="vtmn-flex vtmn-flex-col vtmn-justify-center vtmn-items-end")
```

```python
lista_precios = []

for precio in precios:
    nombre = precio.text.strip()
    lista_precios.append(nombre)

lista_precios
```
